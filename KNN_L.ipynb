```python
# 2-qadam
def euclidean(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
# 3-qadam
def knn_predict(X,y,x_new,k):
    distances = np.linalg.norm(X-x_new,axis=1)
    k_indices = np.argsort(distances)[:k]
    k_labels = np.array(y)[k_indices]
    return np.bincount(k_labels).argmax()
```

```python
# 4-qadam 
def get_distances(X,y,x_new):
    distances = []
    for xi, yi in zip(X,y):
        d = euclidean(xi,x_new)
        distances.append((d,yi))
    return distances    
```

```python
# 7-qadam
from collections import Counter
def vote(k_nearest):
    labels = [label for _,label in k_nearest]
    most_common = Counter(labels).most_common(1)
    return most_common[0][0]
```

```python
# 8-qadam
def knn_predict_full(X, y, x_new, k, labels_map):
    distances = get_distances(X, y, x_new)
    distances = sort_distances(distances)
    k_nearest = get_k_nearest(distances, k)
    result = vote(k_nearest)
    return labels_map[result]
```

```python
# 9 va 10-jptad tayor olingan
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

X = np.array([[1,50],[2,55],[5,80],[6,85]])
y = ["O'ta olmadi","O'ta olmadi","O'tdi","O'tdi"]
x_new = np.array([3, 60])
k = 3

def euclidean(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def get_distances(X, y, x_new):
    return [(euclidean(xi, x_new), yi) for xi, yi in zip(X, y)]

def sort_distances(distances):
    distances.sort(key=lambda x: x[0])
    return distances

def get_k_nearest(distances, k):
    return distances[:k]

def vote(k_nearest):
    labels = [label for _, label in k_nearest]
    return max(set(labels), key=labels.count)

def knn_predict(X, y, x_new, k):
    distances = get_distances(X, y, x_new)
    distances = sort_distances(distances)
    k_nearest = get_k_nearest(distances, k)
    return vote(k_nearest)

my_result = knn_predict(X, y, x_new, k)
print("O'zimizning KNN:", my_result)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
x_new_scaled = scaler.transform([x_new])

model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_scaled, y)
sklearn_result = model.predict(x_new_scaled)[0]
print("Sklearn KNN:", sklearn_result)

if my_result == sklearn_result:
    print("Natijalar bir xil!")
else:
    print("Natijalar farqli!")
```
